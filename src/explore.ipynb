SQL Connection

```python
connection = sqlite3.connect('tmdb.db')

df_movies.to_sql("df_movies", connection, if_exists="replace", index=False)
df_credits.to_sql("df_credits", connection, if_exists="replace", index=False)

cursor = connection.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables:", tables)
```

Join the tables

Merge 2 tables

```python
query = """
    SELECT *
    FROM df_movies
    INNER JOIN df_credits
    ON df_movies.title = df_credits.title;
"""

total_data = pd.read_sql_query(query, connection)
connection.close()
```

```python
total_data = total_data.loc[:, ~total_data.columns.duplicated()]
total_data.head()
```
